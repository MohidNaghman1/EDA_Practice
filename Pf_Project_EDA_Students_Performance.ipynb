### Insights

* Making a new column on the basis of math score


```python
# Checking minimum score in math
min(df['math score'])
```

```python
# Checking maximum score in math
max(df['math score'])
```

```python
# Using Austria's grading system
bin = [0, 50, 63, 79, 89, 100]
# labals = ['0-50', '51-63', '64-79', '80-89', '90-100']
labals = ['Fail(F)', 'Sufficient(D)', 'Satisfactory(C)', 'Good(B)', 'Really good(A)']
df['math_grades'] = pd.cut(df['math score'], bins=bin, labels=labals)
```

```python
df.math_grades.value_counts()
```

* Making a new column on the basis of reading score


```python
bin = [0, 50, 63, 79, 89, 100]
# labals = ['0-50', '51-63', '64-79', '80-89', '90-100']
labals = ['Fail(F)', 'Sufficient(D)', 'Satisfactory(C)', 'Good(B)', 'Really good(A)']
df['reading_grades'] = pd.cut(df['reading score'], bins=bin, labels=labals)
```

```python
df.reading_grades.value_counts()
```

* Making a new column on the basis of writing score


```python
bin = [0, 50, 63, 79, 89, 100]
# labals = ['0-50', '51-63', '64-79', '80-89', '90-100']
labals = ['Fail(F)', 'Sufficient(D)', 'Satisfactory(C)', 'Good(B)', 'Really good(A)']
df['writing_grades'] = pd.cut(df['writing score'], bins=bin, labels=labals)
```

```python
df.writing_grades.value_counts()
```

* Now groupby using gender

```python
# Grouping math grades by using gender
df.groupby('gender')['math_grades'].value_counts()
# Boys are more intelligent in math instead of Girls
```

```python
# Grouping reading grades by using gender
df.groupby('gender')['reading_grades'].value_counts()
# Girls are more intelligent in reading instead of Boys
```

```python
# Grouping math grades by using gender
df.groupby('gender')['writing_grades'].value_counts()
# Girls are more intelligent in writing instead of Boys
```
